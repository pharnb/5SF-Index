```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import pymongo
```

```python
# Initialize PyMongo to work with MongoDBs
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
# Define database and collection
db = client.FiveSF
collection = db.episodes
```

```python
# URL of page to be scraped, excluding year
url = 'https://www.imdb.com/title/tt7370554/episodes?year='

# Retrieve page with the requests module
# loop over years range for all episodes and append to url
for year in range(2008,2021):
    response = requests.get(url + str(year))
    # Create BeautifulSoup object; parse with 'lxml'
    soup = BeautifulSoup(response.text, 'lxml')
    # Retrieve the parent divs for all episodes
    results = soup.find_all('div', class_='list_item')
    # print(results)

    # loop over results to get episode list
    for result in results:
        try :
            # scrape the title of the 5SF
            title = result.a['title']

            # scrape the date listed
            date = result.find('div', class_='airdate').text

            # remove leading/trailing spaces from strings
            title = title.strip()
            date = date.strip()

            # scrape the datetime
            # datetime = result.find('span', class_='article-item__date')['data-date'] 

            # get only the date from the datetime
            # date = datetime.split('T')[0]

            # print article data
            print('-----------------')
            print(title)
            print(date)
            # # print(date)

            # Dictionary to be inserted into MongoDB
            post = {
                'title': title,
                # 'subheader': subheader,
                'date': date,
            }

            # Insert dictionary into MongoDB as a document
            collection.insert_one(post)
        except :
            pass
```

```python
# Display the MongoDB records created above
fivesf = db.episodes.find()
for episode in fivesf:
    print(episode)
```
